```python
import sqlite3
import pandas as pd
import os
conn = sqlite3.connect('mailchimp_test.db')
cursor = conn.cursor()
```

```python
query ="""
SELECT * 
FROM CONTACT_LIST 
WHERE PHONE IS NOT NULL
    AND MAILCHIMP_IMPORT IS NOT NULL
    AND CLICKUP_IMPORT IS  NULL
LIMIT 100
"""
```

```python
#sqalchemy query execution
top = pd.read_sql_query(query, conn)
top.head()
```

```python
#CONVERT THE EMAIL COLUMN INTO A STRING TO SQL PROCESSING
email_list = top['EMAIL']
email_string= '('
for i in range(len(email_list)):
    if i < len(email_list)-1:
        email_string = email_string + f"'{email_list[i]}',"
    else:
        email_string = email_string + f"'{email_list[i]}'"
email_string= email_string + ")"
print(email_string)
```

```python
#VALIDATION SQL THAT USES THE EMAIL_STRING SQL TO VERIFY UPDATES
filter = pd.read_sql_query(f"SELECT * FROM CONTACT_LIST WHERE EMAIL IN {email_string} limit 10", conn)
filter
```

```python
#CONVERT THE EMAIL_LIST INTO A LIST OF TUPLES FOR EXECUTION BY EXECUTMANY METHOD
stmt = [(i,) for i in email_list]
print(stmt[:10])
```

```python
#EXECTUMANY
cursor.executemany(f"UPDATE CONTACT_LIST SET CLICKUP_IMPORT = 'Y' WHERE EMAIL = ?",stmt)
conn.commit()
```

```python
#pull contacts that were updated
contacts = pd.read_sql_query(f"SELECT * FROM CONTACT_LIST WHERE EMAIL IN {email_string}", conn)
contacts.head()

```
