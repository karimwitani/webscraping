```python
import sqlite3
import pandas as pd
import os
conn = sqlite3.connect('mailchimp.db')
cursor = conn.cursor()
```

```python
imported = pd.read_csv('clickup_imports.csv')
imported.head()
```

```python
imported[' Task Name']
```

```python
#CONVERT THE EMAIL COLUMN INTO A STRING TO SQL PROCESSING
name_list = imported[' Task Name']
name_string= '('
for i in range(len(name_list)):
    if i < len(name_list)-1:
        name_string = name_string + f"'{name_list[i]}',"
    else:
        name_string = name_string + f"'{name_list[i]}'"
name_string= name_string + ")"
print(name_string)
```

```python
#BASIC SEARCH OF NAMES IN CLICKUP WITHIN THE DB
filter = pd.read_sql_query(f"SELECT * FROM CONTACT_LIST WHERE FULL_NAME IN {name_string}", conn)
filter
```

```python
#NAMES IN CLICKUP NOT YET MARKED AS PULLED IN CLICKUP
filter = pd.read_sql_query(f"SELECT * FROM CONTACT_LIST WHERE FULL_NAME IN {name_string} AND CLICKUP_IMPORT IS NULL", conn)
filter
```

```python
#CONVERT THE NAME_LIST INTO A LIST OF TUPLES FOR EXECUTION BY EXECUTMANY METHOD
stmt = [(i,) for i in name_list]
print(stmt[:10])
```

```python
#EXECTUMANY
cursor.executemany(f"UPDATE CONTACT_LIST SET CLICKUP_IMPORT = 'Y' WHERE FULL_NAME = ?",stmt)
conn.commit()
```

```python
#VALIDATION THAT WE MARKED THEM ALL AS IMPORTED INTO CLICKUP
filter = pd.read_sql_query(f"SELECT * FROM CONTACT_LIST WHERE FULL_NAME IN {name_string} AND CLICKUP_IMPORT IS NULL", conn)
filter
```
